```python
def plot(y_true, y_pred, normalize = False, border = 100):
    fig = plt.figure(figsize=(18,12))
    # set up subplot grid
    gridspec.GridSpec(2,2)
    
    #Normalization of real and predicted values to 1
    if normalize:
        border = 1
        y_true /= max(copy.copy(y_true))
        y_pred /= max(copy.copy(y_pred))
    
    #Big subplot with diagonal plot
    plt.subplot2grid((2,3), (0,0), colspan=2, rowspan=2)
    plt.xlim(0, border)
    plt.ylim(0, border)
    plt.scatter(y_true, y_pred, c = 'k', s = 10, marker = 's')
    plt.plot([0, border], [0, border], c = 'red', alpha = 0.6)
    plt.plot([0, border * 1/1.15], [0, border], color = 'black', alpha = 0.6, linewidth = 2, linestyle = '--')
    plt.plot([0, border], [0, border * 0.85], color = 'black', alpha = 0.6, linewidth = 2, linestyle = '--')
    plt.xlabel('Реальные значения', fontsize = 20)
    plt.ylabel('Прогнозные значения', fontsize = 20)
    plt.tick_params(labelsize = 20)
    plt.grid()
    
    #Subplot with histogramms for real and predicted values
    plt.subplot2grid((2,3), (0,2))
    plt.hist(y_true, alpha = 0.5, bins = 100, label = 'Реальные значения')
    plt.hist(y_pred, color = 'red', alpha = 0.5, bins = 100, label = 'Прогноз')
    plt.title('Распределения прогнозного параметра', fontsize = 16)
    plt.tick_params(labelsize = 16)
    plt.legend(fontsize = 16)
    
    #Subplot with difference histogramm of real and predicted values
    plt.subplot2grid((2,3), (1,2))
    plt.hist(y_true - y_pred, color = 'blue', alpha = 0.75, bins = 100)
    plt.title('Распределение отклонения прогноза от реальности', fontsize = 16)
    plt.tick_params(labelsize = 16)
```

```python
    def __encirclement_pairs(self, df_coord):
        '''
        Function to find all pairs of wells in gives radius
        :param df_coord dataframe with coordinates of all wells
        '''
        df_neib = pd.DataFrame(columns=['well', 'neighbours', 'weights'])
        all_wells = list(set(df_coord['Скважина']))
        coords = df_coord[['Координата X', 'Координата Y']].values
        point_tree = cKDTree(coords)
        for well in tqdm.tqdm(all_wells):
            new_row = {'well':well}
            #Build tree based on coordinates
            #If no coordinates for selected well - put none in dataframe with neigbours
            if df_coord[df_coord['Скважина'] == well].shape[0] < 1:
                new_row.update({'neighbours': None})
                new_row.update({'weights': None})
                df_neib.loc[len(df_neib)] = new_row
                continue
            #Coordinates of considered well
            coord = df_coord[df_coord['Скважина'] == well][['Координата X', 'Координата Y']].values[0]
            tree_inds = point_tree.query_ball_point(coord, self.encirclement_rad)
            neib_wells = []
            weights = [] #Inverse distances
            for i in tree_inds:
                neib_well = df_coord[(df_coord['Координата X'] == coords[i][0]) & \
                                (df_coord['Координата Y'] == coords[i][1])]['Скважина'].values[0]
                #Do not add considered well
                if (coord[0] == coords[i][0]) & (coords[i][1] == coord[1]):
                    continue
                neib_wells.append(neib_well)
                #Euclidian distance calculation
                dist = np.sqrt((coords[i][0] - coord[0]) ** 2 + (coords[i][1] - coord[1]) ** 2)
                #Inverse distances serve as weights
                weights.append(1.0 / dist)
            new_row.update({'neighbours': neib_wells})
            new_row.update({'weights': weights})
            df_neib.loc[len(df_neib)] = new_row
            #self.df_neib = self.df_neib.append(new_row, ignore_index=True)
        return df_neib
```
